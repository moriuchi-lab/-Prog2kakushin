ビンゴで遊べるプログラムを作りました

```python
import numpy as np
import random
```

時間がないけど遊びたい人用

```python
N=5
a=np.random.randint(1,41,(N,N))
b=np.random.randint(1,41,(N,N))
print('card A')
print(a)
print('card B')
print(b)
x=str(input('AとBどちらのカードにしますか? '))
count=0
while count!=-1:
  count+=1
  atari=np.random.randint(1,41)
  print(str(count)+'番目に出た数字は '+str(atari))
  a_new=np.where(a==atari,0,a)
  b_new=np.where(b==atari,0,b)
  a=a_new
  b=b_new
  print('card A')
  print(a_new)
  print('card B')
  print(b_new)
  c=np.sum(a_new,axis=0)
  d=np.sum(a_new,axis=1)
  e=np.sum(b_new,axis=0)
  f=np.sum(b_new,axis=1)
  g=np.diag(a_new)
  h=np.diag(np.fliplr(a_new))
  j=np.diag(a_new)
  k=np.diag(np.fliplr(a_new))
  for i in range(N):
    if c[i]==0:
      print('card A BINGOです!')
      count=-1
      y='A'
      break
    elif d[i]==0:
      print('card A BINGOです!')
      count=-1
      y='A'
      break
    elif np.sum(g)==0:
      print('card A BINGOです!')
      count=-1
      y='A'
      break
    elif np.sum(h)==0:
      print('card A BINGOです!')
      count=-1
      y='A'
      break
    if e[i]==0:
      print('card B BINGOです!')
      count=-1
      y='B'
      break
    elif f[i]==0:
      print('card B BINGOです!')
      count=-1
      y='B'
      break
    elif np.sum(j)==0:
      print('card B BINGOです!')
      count=-1
      y='B'
      break
    elif np.sum(k)==0:
      print('card B BINGOです!')
      count=-1
      y='B'
      break

if x==y:
  print('あなたの勝ちです')
else :
  print('あなたの負けです')
```
